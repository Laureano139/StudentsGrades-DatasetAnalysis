### First of all, I will see all the attributes and then gather information (previously provided) about them.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('studentsData.csv')

# I'll use the function head() to print the first 5 rows of the dataset
print(df.head())
```

```python
df.describe()
```

### Attribute values exploration and explanation:

- **StudentID** => A unique identifier assigned to each student (1001 to 3392).
- **Age** => The age of the students ranges from 15 to 18 years.
- **Gender** => Gender of the students, where 0 represents Male and 1 represents Female.
- **Ethnicity** => The ethnicity of the students, coded as follows: 0 => Caucasian; 1 => African American; 2 => Asian; 3 => Other.
- **ParentalEducation** => The education level of the parents, coded as follows: 0 => None; 1 => High School; 2 => Some College; 3 => Bachelor's; 4 => Higher.
- **StudyTimeWeekly** => Weekly study time in hours, ranging from 0 to 20.
- **Absences** => Number of absences during the school year, ranging from 0 to 30.
- **Tutoring** => Tutoring status, where 0 indicates No and 1 indicates Yes.
- **ParentalSupport** => The level of parental support, coded as follows: 0 => None; 1 => Low; 2 => Moderate; 3 => High; 4 => Very High.
- **Extracurricular** => Participation in extracurricular activities, where 0 indicates No and 1 indicates Yes.
- **Sports** => Participation in sports, where 0 indicates No and 1 indicates Yes.
- **Music** => Participation in music activities, where 0 indicates No and 1 indicates Yes.
- **Volunteering** => Participation in volunteering, where 0 indicates No and 1 indicates Yes.
- **GPA** => Grade Point Average on a scale from 2.0 to 4.0, influenced by study habits, parental involvement, and extracurricular activities.

### Target

**GradeClass**: Classification of students' grades based on GPA:
- 0: 'A' (GPA >= 3.5)
- 1: 'B' (3.0 <= GPA < 3.5)
- 2: 'C' (2.5 <= GPA < 3.0)
- 3: 'D' (2.0 <= GPA < 2.5)
- 4: 'F' (GPA < 2.0)

### The attribute "StudentID" is not important for my analysis, so I will get rid of it. 

```python
if 'StudentID' in df.columns:
    df = df.drop('StudentID', axis=1)

print(df.head())
```

```python
# Now, I will loop through all columns to create bar charts to see if there are any outliers/strange values

for column in df.columns:
    if df[column].dtype == 'object':  # Here I checked if the attribute is categorical
        plt.figure(figsize=(10, 6))
        sns.countplot(x=df[column], order=df[column].value_counts().index)
        plt.title(f'{column} distribution')
        plt.xticks(rotation=45)
        plt.show()
    else:
        plt.figure(figsize=(10, 6))
        df[column].value_counts().sort_index().plot(kind='bar')
        plt.title(f'{column} distribution')
        plt.show()
```

### As we can see, the attributes got the correct values, although there are some graphics that are inconclusive in terms of visualization...we will deal with that later. 

```python
missing_values = df.isnull().sum()
print(missing_values)
```

### No missing values, that's great! Now I will show the correlation matrix.

```python
plt.figure(figsize=(12, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

### Important information to retain from the matrix:
1. Strong negative correlation between "Absences" and "GPA", which makes sense
   - The more abscenses, the less GPA (lower grades).
2. Strong correlation between "Absences" and "GradeClass"
   - GradeClass is numbered from 0 to 4, with 4 beeing the worst grade ('F') with the lowest GPA's, so it's correct
3. Moderate correlation between "StudyTimeWeekly" and "GPA"
   - The time spent studying per week has impact in the GPA
  
### Although this informations are important, the "GPA" can be affected by other attributes.

```python
# Possible outliers ?

for column in df.columns:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=df[column])
    plt.title(f'{column} boxplot')
    plt.show()
```

### Given the context and the previous explanation of the dataset, it seems that we are free of outliers.

### Now, I will show more graphs to see if I can conclude any other things from other attributes, using "GPA" as a metric, but first, let's apply the binning technique to "GPA", since the intervals in "GradeClass" are made by different intervals of GPA values:

- 0: 'A' (GPA >= 3.5)
- 1: 'B' (3.0 <= GPA < 3.5)
- 2: 'C' (2.5 <= GPA < 3.0)
- 3: 'D' (2.0 <= GPA < 2.5)
- 4: 'F' (GPA < 2.0)

```python
# Here I defined the intervals for the bins
bins = [0, 2.0, 2.5, 3.0, 3.5, 4.0]
labels = ['F', 'D', 'C', 'B', 'A']

# Applying the binning to the GPA column
df['GPA_Category'] = pd.cut(df['GPA'], bins=bins, labels=labels, right=False)

gpa_distribution = df['GPA_Category'].value_counts().sort_index()
print(gpa_distribution)
```

```python
fig, ax = plt.subplots(1, 3, figsize=(18, 6))

sns.countplot(x='GPA_Category', hue='Sports', data=df, ax=ax[0])
ax[0].set_title('GPA Category by Sports')
ax[0].set_xlabel('GPA Category')
ax[0].set_ylabel('Count')
ax[0].legend(title='Sports Practice', loc='upper left', labels=['No', 'Yes'])

sns.countplot(x='GPA_Category', hue='Music', data=df, ax=ax[1])
ax[1].set_title('GPA Category by Music')
ax[1].set_xlabel('GPA Category')
ax[1].set_ylabel('Count')
ax[1].legend(title='Music Practice', loc='upper left', labels=['No', 'Yes'])

sns.countplot(x='GPA_Category', hue='Volunteering', data=df, ax=ax[2])
ax[2].set_title('GPA Category by Volunteering')
ax[2].set_xlabel('GPA Category')
ax[2].set_ylabel('Count')
ax[2].legend(title='Volunteering Practice', loc='upper left', labels=['No', 'Yes'])

plt.tight_layout()
plt.show()
```

### As we can see, the extracurricular activities don't seem to have a direct impact in the grades, but it can impact (positivetly) the happiness and well being of the student, so, in my opinion, it's completely good for the student being enrolled in some extracurricular activity. 

```python
# Parental Support - GPA analysis

plt.figure(figsize=(10, 6))
sns.barplot(x='ParentalSupport', y='GPA', data=df, errorbar=None)
plt.title('Parental support impact on GPA')
plt.xlabel('Parental Support level')
plt.ylabel('GPAs mean')
plt.show()
```

### As we can see, the more parental support, the higher the GPAs mean. Although it seems like a small factor, sometimes, it can make a difference.

```python
# Parents education level impact on GPA

plt.figure(figsize=(10, 6))
sns.barplot(x='ParentalEducation', y='GPA', data=df, errorbar=None)
plt.title('Parents education level impact on GPA')
plt.xlabel('Parents education level')
plt.ylabel('GPAs mean')
plt.xticks(ticks=[0, 1, 2, 3, 4], labels=['None', 'High School', 'Some College', 'Bachelors', 'Higher'])
plt.show()
```

### As we can see, it seems that the parents education level doesn't have an impact on the GPA, as the mean for every category is, practically the same.

### Now, for fun, let's see who has the best GPA mean, boys vs. girls!

```python
plt.figure(figsize=(8, 6))
sns.barplot(x=df['Gender'].replace({0: 'Male', 1: 'Female'}), y='GPA', data=df)
plt.title('GPAs mean by sex')
plt.xlabel('Sex')
plt.ylabel('GPAs mean')
plt.show()
```
